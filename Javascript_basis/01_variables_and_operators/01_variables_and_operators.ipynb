### Inleiding

In de vorige secties heb je geleerd hoe je webpagina's structureert met HTML en ze opmaakt met CSS. De volgende stap is om de webpagina *interactief* te maken, en dat is precies waar JavaScript voor dient.

### Hoe JavaScript-code uitvoeren

Alle JavaScript die we in het grootste deel van de Foundations-cursus zullen schrijven, wordt uitgevoerd via de browser. Latere lessen in Foundations en het NodeJS-pad laten je zien hoe je JavaScript buiten de browseromgeving kunt uitvoeren. Dit betekent dat we de browser onze code laten uitvoeren, zelfs als deze afkomstig is uit een bestand dat we zelf hebben gemaakt.

Buiten deze lessen moet je voorlopig altijd standaard je JavaScript in de browser uitvoeren, tenzij anders aangegeven, anders kun je onverwachte fouten tegenkomen.

De eenvoudigste manier om te beginnen is door een HTML-bestand te maken met de JavaScript-code erin. Gebruik de VS Code-sneltoets <kbd>!</kbd> + <kbd>TAB</kbd> om het basis HTML-skelet in een bestand op je computer te maken. Zorg ervoor dat je de `<script>`-tag toevoegt:

```html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Document</title>
</head>
<body>

  <script>
    // Je JavaScript komt hier!
    console.log("Hello, World!")
  </script>

</body>
</html>
```

Sla dit bestand op en open het in een webbrowser en open vervolgens de console van de browser:

1. Klik met de rechtermuisknop op de lege webpagina.
1. Klik op "Inspecteren" of "Element inspecteren" om de ontwikkelaarstools te openen.
1. Zoek en selecteer het tabblad "Console", waar je de uitvoer van onze `console.log`-instructie zou moeten zien.

`console.log()` is het commando om iets af te drukken naar de ontwikkelaarsconsole in je browser. Je kunt dit gebruiken om de resultaten van de volgende artikelen en oefeningen naar de console te printen. We moedigen je aan om alle voorbeelden in deze en toekomstige lessen mee te coderen.

Een andere manier om JavaScript in een webpagina op te nemen is via een extern script. Dit lijkt erg op het koppelen van externe CSS-bestanden aan je website.

```html
<script src="javascript.js"></script>
```

JavaScript-bestanden hebben de extensie `.js`, vergelijkbaar met `.css` voor stylesheets. Externe JavaScript-bestanden worden gebruikt voor meer complexe scripts.

We hebben ons bestand `javascript.js` genoemd, maar we hadden elke naam kunnen kiezen zoals `mijn-script.js` of zelfs geen naam `.js`. Wat echt belangrijk is, is de `.js` extensie.

### Variabelen

Dit zijn de bouwstenen van elk programma. Je kunt variabelen zien als "opslagcontainers" voor gegevens in je code.

![Variable Box Illustration](https://cdn.statically.io/gh/TheOdinProject/curriculum/d39eaf2ca95e80705f703bb218216c10508f5047/foundations/javascript_basics/fundamentals-1/imgs/00.png)

Je kunt variabelen declareren met het sleutelwoord `let`. Laten we het proberen!

```javascript
let name = "John";
let surname = "Doe";

console.log(name);
console.log(surname);
```

Wat zal de `console.log` afdrukken? Probeer het uit!

Je kunt variabelen ook opnieuw toewijzen:

```javascript
let age = 11;
console.log(age); // geeft 11 weer in de console

age = 54;
 
console.log(age); // wat wordt er nu weergegeven?
```

Let op het ontbreken van `let` op regel 4 - dat is niet nodig omdat de variabele eerder al is *gedeclareerd* en we deze hier alleen opnieuw toewijzen!

Opnieuw toewijzen is handig, maar wat als we *niet* willen dat dit gebeurt? Bijvoorbeeld, we hebben misschien een *constante* `pi` die nooit opnieuw hoeft te worden toegewezen. Dit kunnen we bereiken met het sleutelwoord `const`.

```javascript
const pi = 3.14;
pi = 10;

console.log(pi); // Wat wordt er weergegeven?
```

Je intuïtie zegt misschien dat `3.14` wordt weergegeven. Probeer het!

Er wordt een foutmelding weergegeven. Het komt niet eens bij de `console.log`! Je vraagt je misschien af waarom we een fout in onze code zouden *willen*. Eerlijk gezegd zijn fouten ontzettend handig om ons te vertellen wat er mis is met onze code en precies waar het probleem zit. Zonder fouten zou onze code nog steeds niet doen wat we willen, maar het zou een enorme klus zijn om te achterhalen wat er mis is!

Samengevat zijn er twee manieren om een variabele te declareren:

- `let`, die we kunnen hergebruiken.
- `const`, die we **niet** kunnen hergebruiken en die een fout geeft als we het proberen.

Er is ook een derde manier, `var`, die oorspronkelijk werd gebruikt om variabelen in JavaScript te declareren. `var` lijkt op `let` in die zin dat variabelen op deze manier opnieuw kunnen worden toegewezen, maar het heeft andere eigenaardigheden die zijn opgelost toen de taal `let` en `const` introduceerde. Over het algemeen wordt het niet meer gebruikt. Toch zul je waarschijnlijk code tegenkomen die `var` gebruikt, dus het is handig om te weten dat het bestaat.

### Getallen

Getallen zijn de bouwstenen van programmeerlogica! Het is zelfs moeilijk om een nuttige programmeertaak te bedenken waarbij niet minstens een beetje wiskunde komt kijken... dus weten hoe getallen werken is uiteraard erg belangrijk. Gelukkig is het ook vrij eenvoudig.

Als je naar school bent geweest, zul je het concept waarschijnlijk niet moeilijk vinden. Bijvoorbeeld, de wiskundige uitdrukking `(3 + 2) - 76 * (1 + 1)` is ook geldige JavaScript. Als je dat in een `console.log` zet, zal het de uitdrukking evalueren en het juiste getal weergeven. Probeer het!

Probeer de volgende oefeningen door code toe te voegen aan een script-tag in je HTML-bestand:
