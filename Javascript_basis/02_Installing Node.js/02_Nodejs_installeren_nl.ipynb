### Inleiding

Node.js is een JavaScript-runtimeomgeving waarmee je JavaScript buiten je webbrowser kunt uitvoeren. We hebben dit nodig voor enkele oefeningen in de komende lessen. Om te beginnen zijn er enkele vereiste tools die we moeten installeren voordat we Node op je systeem kunnen zetten.

We gaan Node installeren met behulp van `nvm` (Node Version Manager), omdat het eenvoudig is om hiermee Node-versies te wijzigen en te upgraden. Er is ook een andere tool genaamd `npm` die we later zullen gebruiken om verschillende bibliotheken en tools te installeren die in JavaScript-omgevingen worden gebruikt. Deze twee worden vaak verward, dus lees aandachtig!

Node is ook heel eenvoudig te installeren met nvm, dus dit zou snel moeten gaan :)

### NVM installeren

<details markdown="block">

<summary class="dropDown-header">Installatie op Linux</summary>

#### Stap 0: Vereisten

Om nvm correct te installeren, heb je `curl` nodig. Voer het onderstaande commando uit om `curl` te installeren:

```bash
sudo apt install curl
```

Opmerking: Mogelijk moet je de Ubuntu-pakketlijsten bijwerken naar de nieuwste versie voordat de installatie van Curl voltooid kan worden. Voer in dat geval het onderstaande commando uit:

```bash
sudo apt update && sudo apt upgrade
```

#### Stap 1: NVM downloaden en installeren

Voer dit commando uit om `nvm` te installeren:

```bash
curl -o- https://raw.githubusercontent.com/nvm-sh/nvm/v0.40.3/install.sh | bash
```

#### Stap 2: NVM initialiseren

In de terminal zouden instructies moeten verschijnen over hoe je `nvm` initialiseert. Zo niet (of als je het niet uit de terminal wilt kopiëren), voer dan deze commando's uit:

```bash
export NVM_DIR="$([ -z "${XDG_CONFIG_HOME-}" ] && printf %s "${HOME}/.nvm" || printf %s "${XDG_CONFIG_HOME}/nvm")"
[ -s "$NVM_DIR/nvm.sh" ] && \. "$NVM_DIR/nvm.sh" # Dit laadt nvm
```

Je kunt controleren of `nvm` is geïnstalleerd door het volgende commando uit te voeren:

```bash
command -v nvm
```

Als dit `nvm: command not found` retourneert, sluit dan de terminal en open deze opnieuw.

</details>

<details markdown="block">

<summary class="dropDown-header">Installatie op macOS</summary>
  
Op macOS 10.15 en hoger is de standaard shell nu zsh. Tijdens de installatie zoekt nvm naar een `.zshrc`-bestand in je thuismap. Standaard bestaat dit bestand niet, dus we moeten het aanmaken.

Om het `.zshrc`-bestand aan te maken en de nvm-installatie te starten, voer je de volgende commando's uit:

```bash
touch ~/.zshrc
```

```bash
curl -o- https://raw.githubusercontent.com/nvm-sh/nvm/v0.40.1/install.sh | bash
```

Herstart je terminal, of kopieer en plak het volgende in je terminal en druk op Enter:

```bash
export NVM_DIR="$HOME/.nvm"
[ -s "$NVM_DIR/nvm.sh" ] && \. "$NVM_DIR/nvm.sh" # Dit laadt nvm
[ -s "$NVM_DIR/bash_completion" ] && \. "$NVM_DIR/bash_completion" # Dit laadt nvm bash_completion
```

Test je nvm-installatie door het volgende uit te voeren:

```bash
nvm --version
```

Voor meer informatie, lees over [installeren en updaten via NVM's GitHub](https://github.com/nvm-sh/nvm#installation-and-update).

</details>

### Node installeren

Nu we `nvm` hebben geïnstalleerd, kunnen we Node installeren.

#### Stap 1: Installatie

Voer uit:

```bash
nvm install --lts
```

Dit installeert de meest recente stabiele versie van Node met 'long-term support' (LTS), en je zult veel output in de terminal zien. Als alles goed is gegaan, zie je ergens in de regels iets als het volgende (met de X'en vervangen door echte cijfers):

```bash
Downloading and installing Node vXX.xx.x...
```

Zo niet, sluit dan de terminal, open hem opnieuw en voer `nvm install --lts` opnieuw uit.

#### Stap 2: Node-versie instellen

We moeten `nvm` vertellen welke versie van Node gebruikt moet worden als we het `node`-commando uitvoeren. Dat is eenvoudig; voer het volgende commando uit:

```bash
nvm use --lts
```

We hebben `nvm` nu verteld om de meest recente LTS-versie van Node te gebruiken die op onze computer is geïnstalleerd. Je **moet** de LTS-versie van Node gebruiken om incompatibiliteit met pakketten die we in toekomstige lessen zullen installeren te voorkomen. De LTS-versie van Node krijgt dertig maanden ondersteuning na de eerste release. Het is stabieler en compatibeler met verschillende pakketten dan een niet-LTS-versie van Node.

Als je nu `node -v` uitvoert, zou je `vXX.xx.x` of iets dergelijks moeten zien (met de X'en vervangen door echte cijfers).

Node installeren zorgt er ook voor dat npm wordt meegeïnstalleerd. Voer `npm -v` uit en je zou, net als hierboven, `XX.xx.x` of iets dergelijks moeten zien.

Als je deze berichten ziet, heb je Node en npm succesvol geïnstalleerd!

### De Node-console gebruiken

Voor het gemak biedt Node een interactieve console waarmee je JavaScript-code direct in je terminal kunt uitvoeren en bewerken, vergelijkbaar met IRB voor Ruby. Dit is handig om snel kleine stukjes code te testen of te debuggen zonder telkens de browser te openen.

Om de Node-console te starten, open je je terminal en typ je `node`. Typ `.exit` om de console te verlaten.
